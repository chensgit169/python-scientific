# Curvilinear Coordinate Systems

```python
import sympy as sp
```

```python
from sympy.abc import x, y, z, r, theta, phi
```

```python
# Transformation from polar coordinates to Cartesian coordinates
x_cart = r * sp.cos(theta)
y_cart = r * sp.sin(theta)

# Jacobian matrix for polar coordinates
J_polar = sp.Matrix([[sp.diff(x_cart, r), sp.diff(x_cart, theta)],
                      [sp.diff(y_cart, r), sp.diff(y_cart, theta)]])
```

```python
J_polar
```

```python
# Transformation from spherical coordinates to Cartesian coordinates
z_cart = r * sp.cos(theta)
x_spherical = r * sp.sin(theta) * sp.cos(phi)
y_spherical = r * sp.sin(theta) * sp.sin(phi)

# Jacobian matrix for spherical coordinates
J_spherical = sp.Matrix([[sp.diff(x_spherical, r), sp.diff(x_spherical, theta), sp.diff(x_spherical, phi)],
                          [sp.diff(y_spherical, r), sp.diff(y_spherical, theta), sp.diff(y_spherical, phi)],
                          [sp.diff(z_cart, r), sp.diff(z_cart, theta), sp.diff(z_cart, phi)]])
```

```python
J_spherical
```

```python
# Metric tensor g_ab for polar coordinates
g_polar = sp.Matrix([[1, 0],
                     [0, r**2]])

# Metric tensor g_ab for spherical coordinates
g_spherical = sp.Matrix([[1, 0, 0],
                          [0, r**2, 0],
                          [0, 0, r**2 * sp.sin(theta)**2]])
```

```python
# Function to calculate Christoffel symbols
def christoffel_symbols(g, *args):
    n = len(args)  # Number of dimensions
    
    # Initialize Christoffel symbols tensor
    Gamma = [[[None for k in range(n)] for j in range(n)] for i in range(n)]  # do not use Gamma = [[[None] * n]*n]*n
    
    g_inv = g.inv()  # Inverse of the metric tensor

    # Loop over all indices to compute the Christoffel symbols
    for k, x1 in enumerate(args):
        for i, x2 in enumerate(args):
            for j, x3 in enumerate(args):
                _gamma = sum(1/2 * g_inv[k, l] * (
                    sp.diff(g[i, j], x1) +
                    sp.diff(g[i, l], x2) -
                    sp.diff(g[l, j], x3)) for l in range(n))
                Gamma[k][i][j] = _gamma
                
    return Gamma
```

```python
Gamma_polar = christoffel_symbols(g_polar, r, theta)

# Output results
sp.init_printing(use_latex=True)

print("Christoffel Symbols for Polar Coordinates:")
display(Gamma_polar)
```

```python
# Calculate Christoffel symbols for polar and spherical coordinates
Gamma_polar = christoffel_symbols(g_polar, r, theta)
Gamma_spherical = christoffel_symbols(g_spherical, r, theta, phi)

# Output results
sp.init_printing(use_latex=True)

# Print Jacobian for polar coordinates
print("Jacobian for Polar Coordinates:")
display(J_polar)

# Print Jacobian for spherical coordinates
print("Jacobian for Spherical Coordinates:")
display(J_spherical)

# Print metric tensor for polar coordinates
print("Metric Tensor g_ab for Polar Coordinates:")
display(g_polar)

# Print metric tensor for spherical coordinates
print("Metric Tensor g_ab for Spherical Coordinates:")
display(g_spherical)

# Print Christoffel symbols for polar coordinates


# Print Christoffel symbols for spherical coordinates
print("Christoffel Symbols for Spherical Coordinates:")
display(Gamma_spherical)
```
